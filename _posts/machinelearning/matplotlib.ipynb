## 绘图
- 基础用法

```python
from matplotlib import pyplot as plt
import random
range
x=range(0,120)
# 列表解析语法
y=[random.randint(20,35) for i in range(120)]
#figure=plt.figure()
```

- 设置图片大小

```python
plt.figure(figsize=(20,8),dpi=80)
```

```python
xtick_labels=["10点{}分".format(i) for i in range(60)]
xtick_labels+=["11点{}分".format(i) for i in range(60)]
plt.xticks(list(x)[::30],xtick_labels[::3],rotation=45,fontproperties='simHei')

#添加描述信息
plt.xlabel('时间',fontproperties='simHei')
# 绘制网格

plt.grid()

plt.plot(x,y)
currentFigure= plt.gcf()
plt.show()
```

- 保存图片（保存为svg格式，在图片放大的时候不会产生锯齿）
  如果plt.savefig()在plt.show()之后调用，会保存空白图片，因为show之后，又创建了一个空白的坐标系。
  解决办法是：
  - 在show之前调用
  - 在调用show之前，获取当前图像后，然后保存


```python

currentFigure.savefig('test1.png')
```
